```python
import pandas as pd
import numpy as np
```

### Rating dataset 

```python
rating_df = pd.read_csv("../anime-data/rating.csv")
```

```python
print(rating_df.head())
```

### Predict the -1 in rating

```python
# user id 42653, has invalid data, we will ignore them

rating_df = rating_df[
  (rating_df['rating'] > 0) &
  (rating_df["user_id"] != 42653)
]
```

### Anime dataset

```python
anime_df = pd.read_csv("../anime-data/anime.csv")
```

```python
r_df = rating_df.reset_index().pivot(
  index='user_id', columns='anime_id', values='rating',
).fillna(0)
```

```python
r_df
```

```python
r = r_df.to_numpy()
```

### SVD

```python
from scipy.sparse.linalg import svds
```

```python
rating_mean = np.mean(r, axis=1)
r_demeaned = r - rating_mean.reshape(-1, 1)
```

```python
U, sigma, Vt = svds(r_demeaned, k=20)
```

```python
sigma = np.diag(sigma)
```

```python
preds = np.dot(np.dot(U, sigma), Vt) + rating_mean.reshape(-1, 1)
```

```python
pred_df = pd.DataFrame(preds, columns=r_df.columns)
```

```python
pred_df
```

```python
def recommend_movie(pred_df, user_id, movie_df, origin_rating_df, num=5):
    user_index = user_id - 1
    sorted_user_preds = pred_df.iloc[user_index].sort_values(ascending=False)

    # existing rating dataset, for comparison
    existing_user_ratings = origin_rating_df[origin_rating_df["user_id"] == user_id]
    existing_ratings_df = existing_user_ratings.merge(
        movie_df, how="left", left_on="anime_id", right_on="anime_id"
    ).sort_values(["rating"], ascending=False)

    print(f"User {user_id} has already rated {existing_ratings_df.shape[0]} movies")

    recommends = (
        (movie_df[~movie_df["anime_id"].isin(existing_user_ratings["anime_id"])])
        .merge(
            sorted_user_preds.reset_index(),
            how="left",
            left_on="anime_id",
            right_on="anime_id",
        )
        .rename(columns={user_index: "Predictions"})
        .sort_values("Predictions", ascending=False)
        .iloc[:num]
    )

    return existing_ratings_df, recommends
```

```python
anime_df = anime_df.rename(
  columns={
    "rating": "old_rating"
  }
)

already_rated, user_preds = recommend_movie(pred_df=pred_df, user_id=426, movie_df=anime_df, origin_rating_df=rating_df, num=10,)
```
